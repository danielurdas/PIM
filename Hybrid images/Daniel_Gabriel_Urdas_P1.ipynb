## Cargamos las imagenes

```python
# Importamos las librerias necesarias
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from scipy import ndimage
from scipy import fftpack
import time
%matplotlib inline
```

```python
# Cargamos las imagenes
img_human = Image.open("human.png") 
img_cat = Image.open("cat.png") 

# Redimensionamos las imagenes
size1=int(round(img_human.size[0]*(4/3.))); 
size2=int(round(img_human.size[1]*(4/3.))); 
img_human=img_human.resize((size1,size2), Image.ANTIALIAS) 

# Hacemos un crop de la imagen del hombre por tal de tener la cara alineada con la del gato
left = 50
top = 50
right = 51+img_cat.size[0]-1
bottom = 51+img_cat.size[1]-1
img_human=img_human.crop((left, top, right, bottom))

# Pasamos las imagenes a numpy array y las convertimos a double
img_human_np = np.array(img_human).astype('double')
img_cat_np = np.array(img_cat).astype('double')
```

## Función gauss

###### Función gauss que dada la caracteristica sigma devuelve una matriz bidimensional de medida 6 * sigma + 1 que representa una mascara gaussiana

```python
def gauss(sigma):
    maskSize = 6 * sigma + 1
    mask = np.zeros((maskSize, maskSize))
    for x in range(maskSize):
        for y in range(maskSize):
            mask[x,y] = (1 / (2 * np.pi * sigma**2)) * np.exp(- (((x - (maskSize / 2.))**2 + (y - (maskSize / 2.))**2)/(2*sigma**2))) 
    return mask

mask = gauss(9)
plt.imshow(mask)
plt.axis("off")
print "Suma de los elementos de la matriz: " + str(np.sum(mask))
```

# Dominio del espacio

###### Función filtro_lp_e que dada una imagen, le aplica una mascara gaussiana por tal de eliminar las frecuencias altas. Se hace a traves de la convolución en el dominio del espacio

```python
def filtro_lp_e(img):
    mask = gauss(9)
    img_lp = np.zeros(img.shape)
    # Aplicamos la mascara por cada canal utilizando la convolución
    img_lp[:,:,0] = ndimage.convolve(img[:,:,0],mask, mode='constant')
    img_lp[:,:,1] = ndimage.convolve(img[:,:,1],mask, mode='constant')
    img_lp[:,:,2] = ndimage.convolve(img[:,:,2],mask, mode='constant')
    
    #Una vez convolucionada la imagen, la normalizamos por cada canal para no tener valores negativos
    img_lp[:,:,0] = img_lp[:,:,0] - np.min(img_lp[:,:,0])
    img_lp[:,:,1] = img_lp[:,:,1] - np.min(img_lp[:,:,1])
    img_lp[:,:,2] = img_lp[:,:,2] - np.min(img_lp[:,:,2])
    return img_lp
```

###### Utilizando el filtro pasa bajos en el dominio del espacio, obtenemos la imagen LP del hombre y la imagen HP del gato. A continuacón, fusionamos las dos imagenes por tal de crear la imagen hibrida.

```python
# Calculamos el tiempo que tarda en aplicar el filtro LP a las imagenes
t0 = time.time()
img1_lp = filtro_lp_e(img_human_np)
img2_lp = filtro_lp_e(img_cat_np)
t1 = time.time()
t_esp = t1 - t0

# Obtenemos la imagen HP del gato
img_cat_hp = img_cat_np - img2_lp

# Normalizamos la imagen HP del gato por cada canal
img_cat_hp[:,:,0] = img_cat_hp[:,:,0] - np.min(img_cat_hp[:,:,0])
img_cat_hp[:,:,1] = img_cat_hp[:,:,1] - np.min(img_cat_hp[:,:,1])
img_cat_hp[:,:,2] = img_cat_hp[:,:,2] - np.min(img_cat_hp[:,:,2])

# Fusionamos la imagen Lp del hombre con la imagen HP del gato
img_h_esp = (img1_lp + img_cat_hp) / 2.0

# Mostramos los resultados en una figura
f1 = plt.figure(1) 
plt.subplot(131) 
imgplot2=plt.imshow(np.array(img1_lp).astype('uint8')) 
imgplot2.set_cmap('gray') 
plt.axis("off")
plt.title('Imagen LP', size=16)
plt.subplot(132) 
imgplot2=plt.imshow(np.array(img_cat_hp).astype('uint8')) 
imgplot2.set_cmap('gray') 
plt.axis("off")
plt.title('Imagen HP', size=16)
plt.subplot(133) 
imgplot2=plt.imshow(np.array(img_h_esp).astype('uint8')) 
imgplot2.set_cmap('gray') 
plt.axis("off")
plt.title('Imagen Hibrida', size=16)
plt.gcf().set_size_inches((14,14))
plt.show()
```

###### Función filtro_lp_freq que dada una imagen, le aplica un filtro pasa bajos (gaussiano) por tal de eliminar las frecuencias altas de la imagen(como pueden ser los contornos). Este filtro se le aplica en el dominio frecuencial y el procedimiento es el siguiente: <br><br> 1. Hacemos la transformada de Fourier para los tres canales R G y B de la imagen. <br> 2. Hacemos la transformada de Fourier para la mascara gaussiana. <br> 3. Multiplicamos la matriz de la transformada de Fourier de la imagen con la de la mascara.<br> <bf> 4. Hacemos la inversa de la transformada de Fourier por tal de recuperar la imagen que ahora tendra solo bajas frecuencias  <p>Observaciones: se han utilizado las funciones fft2 y ifft2 port tal de hacer la transformada de Fourier, respectivamente la anti transformada ya que tenemos la imagen en 2D y la función es mas rapida

```python
def filtro_lp_f(img):
    fftsize = 1024
    mask = gauss(9)
    # Pasamos la mascara a numpy array y la convertimos a double
    mask = np.array(mask,'double')
    # Hacemos la transformada de Fourier de la mascara
    mask_fft = fftpack.fft2(mask, (fftsize, fftsize))  

    img_fft = np.zeros((fftsize, fftsize, 3))
    # Hacemos la transformada de Fourier para cada canal de la imagen y el resultado lo multiplicamos con la transformada de la mascara
    img_fft[:,:,0] = fftpack.fft2(img[:,:,0],(fftsize, fftsize)) * mask_fft
    img_fft[:,:,1] = fftpack.fft2(img[:,:,1],(fftsize, fftsize)) * mask_fft
    img_fft[:,:,2] = fftpack.fft2(img[:,:,2],(fftsize, fftsize)) * mask_fft
    
    img_lp = img_fft.copy()
    # Hacemos la anti transformada de Fourier por tal de pasar la imagen en el dominio del espacio
    img_lp[:,:,0] = fftpack.ifft2(img_fft[:,:,0])
    img_lp[:,:,1] = fftpack.ifft2(img_fft[:,:,1])
    img_lp[:,:,2] = fftpack.ifft2(img_fft[:,:,2])
    
    hs=np.floor(mask.shape[0]/2.)
    # Recortamos el area de la imagen original
    im_crop = img_lp[hs:img.shape[0]+hs, hs:img.shape[1]+hs]
    return im_crop
```

# Transformada de Fourier paso a paso

###### A continuación aplicaremos el filtro gaussiano (LP) a una imagen en el dominio de la frecuencia mostrando todos los pasos por tal de verificar que el proceso funciona correctamente. Se ha utilizado la imagen en escala de grises por tal de agilizar el proceso

```python
fftsize = 1024
mask = gauss(9)

# Pasamos la mascara a numpy array y la convertimos a double
mask = np.array(mask,'double')
# Hacemos la transformada de Fourier de la mascara
mask_fft = fftpack.fft2(mask, (fftsize, fftsize))

#Convertimos la imagen a escala de grises
img_human_gray = img_human.convert('L')
#Convertimos la imagen a double
img_human_gray_np = np.array(img_human_gray).astype('double')

img_fft = np.zeros((fftsize, fftsize, 1))
#Aplicamos la transformada de Fourier sobre la imagen
img_fft = fftpack.fft2(img_human_gray_np,(fftsize, fftsize))

# Multiplicamos la transformada de la imagen con la transformada de la mascara por tal de obtener la imagen LP
img_fil_fft = img_fft * mask_fft

img_lp = np.zeros((fftsize, fftsize, 1))
# Hacemos la anti transofrmada de Fourier a la imagen LP por tal de pasarla al dominio del espacio
img_lp = fftpack.ifft2(img_fil_fft)

#Recortamos el area de la imagen original
hs=np.floor(mask.shape[0]/2.)
im_crop = img_lp[hs:img_human_gray_np.shape[0]+hs, hs:img_human_gray_np.shape[1]+hs]

# Mostramos los resultados en una figura
plt.subplot(3,2,1)
plt.imshow(np.array(img_human_gray).astype('uint8'),cmap="gray")
plt.title('Imagen original', size=16)
plt.axis("off")
plt.subplot(3,2,2)
plt.imshow(mask)
plt.title('Filtro gaussiano', size=16)
plt.axis("off")
plt.subplot(3,2,3)
plt.imshow(fftpack.fftshift(np.log(1+np.abs(img_fft))))
plt.title('Imagen FFT', size=16)
plt.axis("off")
plt.subplot(3,2,4)
plt.imshow(fftpack.fftshift(np.log(1+np.abs(mask_fft))))
plt.title('Filtro FFT', size=16)
plt.axis("off")
plt.subplot(3,2,5)
plt.imshow(fftpack.fftshift(np.log(1+np.abs(img_fil_fft))))
plt.title('Imagen LP en el dominio frecuencial', size=16)
plt.axis("off")
plt.subplot(3,2,6)
plt.imshow(np.array(im_crop).astype('uint8'),cmap="gray")
plt.title('Imagen LP en el dominio espacial', size=16)
plt.axis("off")
plt.gcf().set_size_inches((14,14))
```

###### Utilizando el filtro pasa bajos en el dominio de la frecuencia, obtenemos la imagen LP del hombre y la imagen HP del gato. A continuacón, fusionamos las dos imagenes por tal de crear la imagen hibrida

```python
# Calculamos el tiempo que tarda en aplicar el filtro LP a las imagenes
t0 = time.time()
img1_lp = filtro_lp_f(img_human_np)
img2_lp = filtro_lp_f(img_cat_np)
t1 = time.time()
t_freq = t1 - t0

# Normalizamos la imagen LP del hombre
img1_lp = ((img1_lp - np.min(img1_lp)) / (np.max(img1_lp) - np.min(img1_lp))) *255
# Obtenemos la imagen HP del gato restando de la imagen original la imagen LP por tal de efliminar las bajas frecuencias
img2_hp = img_cat_np - img2_lp
#Normalizamos la imagen HP del gato
img2_hp = ((img2_hp - np.min(img2_hp)) / (np.max(img2_hp) - np.min(img2_hp))) *255
#Fusionamos la imagen LP del hombre con la imagen HP del gato
img_h = (img1_lp + img2_hp) / 2.0

#Mostramos los resultados en una figura
f1 = plt.figure(1) 
plt.subplot(131) 
imgplot2=plt.imshow(np.array(img1_lp).astype('uint8')) 
imgplot2.set_cmap('gray') 
plt.axis("off")
plt.title('Imagen LP', size=16)
plt.subplot(132) 
imgplot2=plt.imshow(np.array(img2_hp).astype('uint8')) 
imgplot2.set_cmap('gray') 
plt.axis("off")
plt.title('Imagen HP', size=16)
plt.subplot(133) 
imgplot2=plt.imshow(np.array(img_h).astype('uint8')) 
imgplot2.set_cmap('gray') 
plt.axis("off")
plt.title('Imagen Hibrida', size=16)
plt.gcf().set_size_inches((14,14))
plt.show()
```
