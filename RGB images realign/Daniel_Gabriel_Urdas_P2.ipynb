## Separación y alineación de los canales RGB

###### Función cross_correlation que dado un patrón y una zona de búsqueda, hace la correlación entre las dos imagenes, por tal de saber el desplazamiento entre las dos. 

```python
def cross_correlation(canal1, canal2, x , y, d1, d2):
    # Selecionamos el patrón de búsqueda del primer canal
    img1 = canal1[x - d1 : x + d1, y - d1 : y + d1]
    # Selecionamos la área de búsqueda del segundo canal
    img2 = canal2[x - d1 - d2 : x + d1 + d2, y - d1 - d2 : y + d1 + d2]
    # Invertimos los valores de blanco de la imagen por tal de no tener problemas a la hora de utilizar la función correlate2d
    img1= 1 - ((img1 - np.min(img1)) / np.max(img1))
    img2= 1 - ((img2 - np.min(img2)) / np.max(img2))
    # Obtenemos la correlación entre las dos imagenes
    corr = correlate2d(img1.astype('float'), img2.astype('float'),  mode='full', boundary='fill', fillvalue=0)
    # Buscamos el máximo valor de la correlación
    max = np.max(corr)
    # Buscamos la posición del valor máximo y calculamos el desplazamiento teniendo en cuenta el centro de la imágen de correación
    for i in range(corr.shape[0]):
        for j in range(corr.shape[1]):
            if corr[i,j] == max:
                # Se calcula el desplazamiento en x, y
                dx = ((corr.shape[0] -1)/ 2) - i
                dy = ((corr.shape[1] -1)/ 2) - j
                break
    return dx, dy, img1, img2, corr
```

###### Mediante la correlación calculamos el desplazamiento entre el canal R y el canal G

```python
print "Deszplazamiento entre R y G"
RGx, RGy, img1, img2, corr = cross_correlation(RGB[:,:,2], RGB[:,:,0], 165, 140, 15, 3)

plt.subplot(1,3,1)
plt.imshow(img1)
plt.title('Canal R - Patron de busqueda', size=14)
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(img2)
plt.title('Canal G - Zona de busqueda', size=14)
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(corr)
plt.title('La correlacion entre las dos zonas', size=14)
plt.axis("off")
plt.gcf().set_size_inches((14,14))

print "Dezplazamiento en x de R con G:", RGx
print "Dezplazamiento en y de R con G:", RGy
```

###### Mediante la correlación calculamos el desplazamiento entre el canal R y el canal B

```python
print "Deszplazamiento entre R y B"
RBx, RBy, img1, img2, corr = cross_correlation(RGB[:,:,2], RGB[:,:,1], 165, 140, 15, 3)

plt.subplot(1,3,1)
plt.imshow(img1)
plt.title('Canal R - Patron de busqueda', size=14)
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(img2)
plt.title('Canal G - Zona de busqueda', size=14)
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(corr)
plt.title('La correlacion entre las dos zonas', size=14)
plt.axis("off")
plt.gcf().set_size_inches((14,14))

print "Dezplazamiento en x de R con B:", RBx
print "Dezplazamiento en y de R con B:", RBy
```

###### Alineamos las dos imagenes después de haber encoenotrado los desplazamientos entre los tres canales

```python
B = RGB[:,:,0]
G = RGB[:,:,1]
R = RGB[:,:,2]
#Desplazamos los pixeles del canal G para coincidir con los del cannal R
G = np.roll(G,-RGx,axis = 0)
G = np.roll(G,-RGy,axis = 1)
#Desplazamos los pixeles del canal B para coincidir con los del cannal R
B = np.roll(B,-RBx,axis = 0)
B = np.roll(B,-RBy,axis = 1)
# Calculamos las dimensiones de las áreas que hay que recortar
deleteX = max(abs(RGx), abs(RBx))
deleteY = max(abs(RGy), abs(RBy))
# Creamos una nueva imagen por tal de eliminar los bordes de la imagen
newImage = np.zeros((np.shape(RGB)[0]-deleteX, np.shape(RGB)[1]-deleteY, np.shape(RGB)[2]))
# Añadimos los canales de la nueva imagen recortando los bordes
newImage[:,:,0] = B[:B.shape[0]-deleteX ,deleteY:B.shape[1]]
newImage[:,:,1] = G[:G.shape[0]-deleteX ,deleteY:G.shape[1]]
newImage[:,:,2] = R[:R.shape[0]-deleteX ,deleteY:R.shape[1]]

# Mostramos la imagen antes y despúes de la alineación
plt.subplot(1,2,1)
plt.imshow(np.array(RGB).astype('uint8'))
plt.title('Imagen sin alineacion', size=18)
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(np.array(newImage).astype('uint8'))
plt.title('Imagen alineada', size=18)
plt.axis("off")
plt.gcf().set_size_inches((20,20))
```

# Aplineación con descomposición piramidal

###### Función gauss para poder hacer el subsampling

```python
def gauss(sigma):
    maskSize = 6 * sigma + 1
    mask = np.zeros((maskSize, maskSize))
    for x in range(maskSize):
        for y in range(maskSize):
            mask[x,y] = (1 / (2 * np.pi * sigma**2)) * np.exp(- (((x - (maskSize / 2.))**2 + (y - (maskSize / 2.))**2)/(2*sigma**2))) 
    return mask
```

###### Aplica una máscara gaussiana a una imagen mediante la convolución. La utilizamos para poder hacer el subsampling

```python
def filtro_lp(img, mask):
    img_lp = np.zeros(img.shape)
    img_lp = ndimage.convolve(img, mask, mode='constant')
    img_lp = img_lp - np.min(img_lp)
    return img_lp
```

###### Función que realiza la descomposición piramidal y calcula la correlación entre la zona de búsqueda y el patrón para cada subsampling

```python
def descomposition(temp, pattern, mask, iterations, samp):
    # Aplicamos la máscara gaussiana y hacemos el subsampling de la zona de búsqueda
    temp_mask = filtro_lp(temp, mask)[::samp, ::samp]
    # Aplicamos la máscara gaussiana y hacemos el subsampling del patrón de búsqueda
    pattern_mask = filtro_lp(pattern, mask)[::samp, ::samp]
    #Obtenemos la correlación entre la zona de búsqueda y el patrón
    corr = correlate2d(pattern_mask.astype('float'), temp_mask.astype('float'), mode='valid')
    
    # Buscamos el máximo valor de la correlación
    max = np.max(corr)
    # Buscamos las cooredenadas del valor máximo de la correlación
    for i in range(corr.shape[0]):
        for j in range(corr.shape[1]):
            if corr[i,j] == max:
                # Guardamos las coordenadas
                dx = ((corr.shape[0] -1)/ 2) - i
                dy = ((corr.shape[1] -1)/ 2) - j
                break
    # Si la iteración llega a 0, hemos llegado al tope de la piramide y devolvemos el desplazamiento
    if (iterations == 0):
        return [dx, dy]
    # Llamamos recursivamente la función de descomposición por tal de hacer un nuevo sampling
    des = descomposition(temp_mask, pattern_mask, mask, iterations-1, samp)
    # Guardamos el desplazamiento del nuevo subsampling           
    des[0] = des[0] + samp * dx
    des[1] = des[1] + samp * dy
    return des
```

###### Realizamos la descomposición piramidal para los tres canales para obtener el desplazamiento

```python
mask = gauss(2)
# Las dimensiones de la zona de búsqueda y del patrón
x = 165
y = 140
d1 = 30
d2 = 5
# Creamos la zona de búsqueda del canal R
tempR = RGB[:,:,2][x - d1 : x + d1, y - d1 : y + d1]
# Creamos el patrón del canal R
patternR = RGB[:,:,2][x - d1 - d2 : x + d1 + d2, y - d1 - d2 : y + d1 + d2]
# Creamos el patrón del canal G
patternG = RGB[:,:,1][x - d1 - d2 : x + d1 + d2, y - d1 - d2 : y + d1 + d2]
# Creamos el patrón del canal B
patternB = RGB[:,:,1][x - d1 - d2 : x + d1 + d2, y - d1 - d2 : y + d1 + d2]
# Calculamos el desplazamiento total para cada uno de los canales
desR = descomposition(tempR, patternR, mask, 4, 2)
desG = descomposition(tempR, patternG, mask, 4, 2)
desB = descomposition(tempR, patternB, mask, 4, 2)
# Calcumos el desplazamiento para los canales G y B respecto al canal R
RGx, RGy = -(desG[0] - desR[0]) , -(desG[1] - desR[1])
RBx, RBy = -(desB[0] - desR[0]) , -(desB[1] - desR[1])

print "Desplazamiento entre R y G:", RGx, RGy
print "Desplazamiento entre R y B:", RBx, RBy
```

###### Alineamos las dos imagenes después de haber encoenotrado los desplazamientos entre los tres canales mediante la descomposición piramidal

```python
B = RGB[:,:,0]
G = RGB[:,:,1]
R = RGB[:,:,2]

#Desplazamos los pixeles del canal G para coincidir con los del cannal R
G = np.roll(G,-RGx,axis = 0)
G = np.roll(G,-RGy,axis = 1)
#Desplazamos los pixeles del canal B para coincidir con los del cannal R
B = np.roll(B,-RBx,axis = 0)
B = np.roll(B,-RBy,axis = 1)
# Calculamos las dimensiones de las áreas que hay que recortar
deleteX = max(abs(RGx), abs(RBx))
deleteY = max(abs(RGy), abs(RBy))
# Creamos una nueva imagen por tal de eliminar los bordes de la imagen
newImage = np.zeros((np.shape(RGB)[0]-deleteX, np.shape(RGB)[1]-deleteY, np.shape(RGB)[2]))
# Añadimos los canales de la nueva imagen recortando los bordes
newImage[:,:,0] = B[:B.shape[0]-deleteX ,deleteY:B.shape[1]]
newImage[:,:,1] = G[:G.shape[0]-deleteX ,deleteY:G.shape[1]]
newImage[:,:,2] = R[:R.shape[0]-deleteX ,deleteY:R.shape[1]]

# Mostramos la imagen antes y despúes de la alineación
plt.subplot(1,2,1)
plt.imshow(np.array(RGB).astype('uint8'))
plt.title('Imagen sin alineacion', size=18)
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(np.array(newImage).astype('uint8'))
plt.title('Imagen alineada', size=18)
plt.axis("off")
plt.gcf().set_size_inches((20,20))
```
